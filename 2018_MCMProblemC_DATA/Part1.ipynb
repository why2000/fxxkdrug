```python
def loadcoor(filename):
    fileinfo = {}
    path = os.path.join('./坐标/',filename)
    with open(path, 'r', encoding='utf-8') as onefile:
        for i, row in enumerate(onefile.readlines()):
            if i == 0:
                continue
            rowsplit = row.split('\t')
            infoneeded = [float(rowsplit[-1].strip('\n').strip('\t').strip(' ')), float(rowsplit[-2].strip('\t').strip(' '))]
            fileinfo[rowsplit[1]] = infoneeded
    return fileinfo
```

```python
def loadpol(filename):
    fileinfo = {}
    path = os.path.join('./人口/',filename)
    with open(path, 'r', encoding='utf-8') as onefile:
        for i, row in enumerate(onefile.readlines()):
            rowsplit = row.split(',')
            infoneeded = int(rowsplit[-1].strip('\n'))
            countyname = rowsplit[0].upper()
#             print(countyname)
            if countyname.endswith(' COUNTY'):
                countyname = countyname[:-7]
#                 print(countyname)
            if countyname in ['RICHMOND']:
                continue
            if countyname.endswith(' CITY'):
                countyname = countyname[:-5]
#                 print(countyname)
            if countyname.startswith('CITY OF '):
                if not countyname[8:] in ['FAIRFAX','FRANKLIN']:
                    countyname = countyname[8:]
#                     print(countyname)
                else:
                    countyname = countyname[8:] +' CITY'
#                     print(countyname)
#             print(countyname+str(namedict[filename][countyname]))
            if countyname not in namedict[filename].keys():
                print(countyname)
                continue
            fileinfo[namedict[filename][countyname]] = infoneeded
    return fileinfo
```

# 半径计算

```python
def caculatela(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371
    return c * r * 1000
```

```python
rmins = []
for x1, y1 in zip(x,y):
    rmin = 1000000
    for x2, y2 in zip(x,y):
        radius = caculatela(x1,y1,x2,y2)
        if radius < rmin and (x1 != x2 or y1 != y2):
            rmin = radius
    rmins.append(rmin)
```

# 改用SIRS

```python
n = 5000
k = 3
p = 0.1
mat = np.zeros((n,n),dtype=np.float16)
for i in range(n):
    for j in range(i,i+k):
        jj = j
        if jj>=n:
            jj = jj%n
        mat[i,jj]=1
        mat[jj,i]=1
for i in range(n):
    for j in range(i,i+k):
        jj = j
        if jj>=n:
            jj = jj%n
        p1 = random.random()
        if p1<p:
            mat[i,jj]=0
            mat[jj,i]=0
            mat[i,i] = float("inf")
            a = np.where(mat[i,:] == 0)
            r = random.randint(0,len(a[0])-1)
            jjj = a[0][r]
            mat[i,jjj]=1
            mat[jjj,i]=1
            mat[i,i]=0
```

# 正向自动机设定

```python
aff_radius = sqrt((max(rmins)**2+min(rmins)**2)/2)
zf = {}
zfdict = {}
for i in range(2010, 2030+1):
    if i == 2010:
        zf[i] = []
        zfdict[i] = {}
        for xi, yi, gid in zip(x,y,gids_raw):
            if gid in yeardict[i].keys():
                zi = yeardict[i][gid]/(poldict[gid]/1000)
                zf[i].append([xi,yi,zi])
                zfdict[i][gid] = zi
    else:
        zf[i] = []
        zfdict[i] = {}
        for x1, y1, gid1 in zip(x,y,gids_raw):
            if gid1 in zfdict[i-1].keys():
                ori_t = zfdict[i-1][gid1]
            else:
                ori_t = 0
            for_res = ori_t
            affs = get_affect(aff_radius, gid1)
            for aff in affs:
                if aff in zfdict[i-1].keys():
                    ori_a = zfdict[i-1][aff]
                else:
                    ori_a = 0
                deltaaff = (ori_a - ori_t*0.60)
                if check_samestate(gid1, aff):
                    deltaaff /= 10
                if len(affs) <= 5:
                    deltaaff /= 5
                if deltaaff > 0:
#                     pass
                    for_res += deltaaff*0.38
                else:
                    for_res += deltaaff*0.58
            zfdict[i][gid1] = for_res
            zf[i].append([x1, y1, for_res])
    to_csv(zf[i], './forward'+str(i)+'.csv')
```
